```python
#Import H2o library 
import h2o
```

```python
#Initialize the server 
h2o.init()
```

```python
#File from which input needs to be taken
url  = "http://h2o-public-test-data.s3.amazonaws.com/smalldata/iris/iris_wheader.csv"
iris = h2o.import_file(url)
```

```python
#Split the DataSet
train,test = iris.split_frame([0.8])
```

```python
#Number of rows in training set  
train.nrows
```

```python
#Number of rows in testing set 
test.nrows
```

```python
#Import Random Forest libraries 
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
# Run the model with training set 
mRF = H2ORandomForestEstimator()
mRF.train(['sepal_len','sepal_wid','petal_len','petal_wid'],'class',train)
```

```python
# Show the properties of Trained Model 
mRF
```

```python
# Run the testing model 
mRF.predict(test)
```

```python
# Performance of the testing model 
mRF.model_performance(test)
```
